```python
from collections import defaultdict
dic = defaultdict(list)
for line in open('Literature for the period between 2004 and 2006.md', 'r'):
    author_date = line.split(') ', 1)[0] + ')'
    dic['author'].append(author_date.split('(')[0][:-1])
    dic['date'].append(author_date.split('(')[1][:-1])
    info = line.split(') ', 1)[1].replace('Travel Time', 'Travel_Time').replace('\x02\x02', '').split()
    dic['Area'].append(info[0])
    dic['Traffic_Parameter'].append(info[1])
    dic['Prediction_Step_min'].append(info[2])
    dic['Prediction_Horizon_steps'].append(info[3])
    dic['data_collection'].append(info[4])
    dic['Methodology_Approach'].append(info[5])
    dic['Problem'].append(info[6])
    dic['Model'].append(info[7])
    if 'Multiple' in info:
        dic['Inputs'].append('Multiple')
    else:
        dic['Inputs'].append('Single')
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(dic)
```

```python
df.to_csv('Literature for the period between 2004 and 2006.csv')
```
